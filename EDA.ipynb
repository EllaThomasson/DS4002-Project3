```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.datasets import mnist

# Step 1: Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Step 2: Preprocess the data
# Normalize the images to be between 0 and 1
x_train = x_train / 255.0
x_test = x_test / 255.0

# Flatten the 28x28 images to a 1D vector of size 784
x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)

# Step 3: Train a logistic regression model
model = LogisticRegression(max_iter=1000, solver='lbfgs', multi_class='multinomial')
model.fit(x_train, y_train)

# Step 4: Predict and evaluate the model
y_pred = model.predict(x_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of Logistic Regression on MNIST: {accuracy * 100:.2f}%')

# Optional: Plot a sample image from the test set
plt.imshow(x_test[0].reshape(28, 28), cmap='viridis')
plt.title(f'Predicted Label: {y_pred[0]}')
plt.show()
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

# Step 1: Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Step 2: Preprocess the data
# Normalize the images to be between 0 and 1
x_train = x_train / 255.0
x_test = x_test / 255.0

# Flatten the 28x28 images to a 1D vector of size 784
x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)

# Step 3: Train a Support Vector Machine model
model = svm.SVC(kernel='linear')  # You can change the kernel, e.g., 'rbf', 'poly', etc.
model.fit(x_train, y_train)

# Step 4: Predict and evaluate the model
y_pred = model.predict(x_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy of SVM on MNIST: {accuracy * 100:.2f}%')

# Optional: Plot a sample image from the test set
plt.imshow(x_test[0].reshape(28, 28), cmap='viridis')
plt.title(f'Predicted Label: {y_pred[0]}')
plt.show()
```
